Описание проекта.  
Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали.  

**Необходимо построить модель, которая предскажет температуру стали.**

Описание этапа обработки.  
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом.  Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Алгоритм обработки:
- Измерить температуру и провести химический анализ.
- Повысить температуру.
- Добавить легирующие материалы и продуть сплав инертным газом.
- Перемешать и снова измерить температуру и провести химический анализ.
- повторять с первого шага пока не будут достигнуты целевой химический состав и ОПТИМАЛЬНАЯ ТЕМПЕРАТУРА плавки.
     
Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки.

Данные содержаться в нескольких файлах. Предсказать надо итоговую температуру плавки, т.е. цель это последнее значение температуры в каждой плавке (файл data_temp_new.csv). Остальные данные - это признаки.  

План анализа:
1. Загрузить данные  
2. Проанализировать полноту, выбросы и анамалии данных  
3. Выделить признаки влияющие на конечную температуру  
4. Создать модели и подобрать гиперпараметры  
5. Обучить и проверить модели  
6. Сделать предсказания на лучшей модели  
7. Проверить лучшую модель на адекватность  
8. Отчет

В данных реактивной мощности в минимальном значении наблюдаем явную аномалию.  
Потребляемая мощность не может быть отрицательной, от заказчика известно, что реактивная мощность это тоже потребление электроэнергии, поэтому генерироваться, да еще в таком количестве, она не может.

Номер партии с аномальным значением реактивной мощности `2116`. Больше отрицательных значений нет. Посмотрим на эту партию подробнее.

```python
df_arc[df_arc['key'] == 2116]
```

Явная ошибка данных. Удалим эту партию из всех датасетов.

```python
def del_key(keys):    
    global df_arc, df_bulk, df_bulk_time, df_gas, df_wire, df_wire_time, df_temp
    for key in keys:
        df_arc = df_arc[df_arc['key'] != key]
        df_bulk = df_bulk[df_bulk['key'] != key]          
        df_bulk_time = df_bulk_time[df_bulk_time['key'] != key]
        df_gas = df_gas[df_gas['key'] != key]
        df_wire = df_wire[df_wire['key'] != key]
        df_wire_time = df_wire_time[df_wire_time['key'] != key]
        df_temp = df_temp[df_temp['key'] != key]

key = [2116]
del_key(key)
df_arc[df_arc['key'] == key[0]]
```

Так как данные о добавках в сталь каждой партии содержаться в одной строке, агрегируем данные затраченной мощности для каждой партии.  
Агрегируем отдельно для активной и реактивной мощности, посчитаем сумму и среднее и, на всякий случай, количество итераций. Может быть это будет дополнительным признаком для модели.  
Также удалим из датасета столбцы с временем, для обучения модели они не нужны.

Жизнь, как всегда, внесла коррективы. Создадим из даты 2 новых признака. Первый - это сумма времени нагревов при каждой итерации. Второй - это общая длительность плавки каждой партии.

Считаем продолжительность нагрева в каждой итерации в каждой плавке. Потом сложим по плавкам.

```python
df_arc['Начало нагрева дугой'] = pd.to_datetime(df_arc['Начало нагрева дугой'], format='%Y%m%d %H:%M:%S')
df_arc['Конец нагрева дугой'] = pd.to_datetime(df_arc['Конец нагрева дугой'], format='%Y%m%d %H:%M:%S')
df_arc['Продолжительность нагрева'] = (df_arc['Конец нагрева дугой'] - df_arc['Начало нагрева дугой']).astype('int64')/1000000000
```

```python
df_arc.head(4)
```

Теперь посчитаем продолжительность каждой плавки.

```python
df_arc_temp = df_arc.groupby('key')[['Начало нагрева дугой', 'Конец нагрева дугой']].agg(['first', 'last'])
df_arc_temp = df_arc_temp[[('Начало нагрева дугой', 'first'), ('Конец нагрева дугой',  'last')]]
df_arc_temp.columns = ['Начало нагрева дугой', 'Конец нагрева дугой']
df_arc_temp['p_time'] = (df_arc_temp['Конец нагрева дугой'] - df_arc_temp['Начало нагрева дугой']).astype('int64')/1000000000
df_arc_temp['key'] = df_arc_temp.index
df_arc_temp.reset_index(drop=True, inplace=True)
df_arc_temp = df_arc_temp[['key', 'p_time']]
df_arc_temp.head(4)
```

Посчитаем сумму и среднее для активной и реактивной мощности, количество итераций и продолжжительность нагрева.

```python
df_arc_group = df_arc.groupby('key').agg({'Активная мощность':['sum', 'median'],
                                          'Реактивная мощность':['sum', 'median', 'count'], 
                                          'Продолжительность нагрева':['sum']})
df_arc_group.columns = ['ap_sum', 'ap_mean', 'rp_sum', 'rp_mean', 'p_count', 'p_sum']
df_arc_group['key'] = df_arc_group.index
df_arc_group = df_arc_group[['key', 'ap_sum', 'ap_mean', 'rp_sum', 'rp_mean', 'p_count', 'p_sum']] 
df_arc_group.reset_index(drop=True, inplace=True)
df_arc_group.head(4)
```

К получившийся таблице признаков добавить признак - общую продолжительность плавки.

```python
df_arc_group = df_arc_group.merge(df_arc_temp, how='left', on='key')
```
